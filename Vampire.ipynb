```python
def ChooseKilled(prev_values):
    n = len(prev_values) + 1
    survivors = 0
    opt_poisoned = 0
    for k in range(1, int(1 * n ** 0.5) + 3):
        s = (n - k) / n * (k + prev_values[n - k - 2])
        if survivors < s:
            opt_poisoned = k
        survivors = max(s, survivors)
    return([survivors, opt_poisoned])
prev_values = [0, 0.5, 2/3, 9 / 8]
    
ChooseKilled(prev_values)
    
```

```python
prev_values = [0, 0.5, 2/3]
killedz = [0, 1, 1]
for k in range(100000):
    prev_values.append(ChooseKilled(prev_values)[0])
    killedz.append(ChooseKilled(prev_values)[1])
prev_values, killedz
pd.DataFrame(killedz).plot()
pd.DataFrame(prev_values).plot()
```
